**K-means clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 1: Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Step 2: Visualize the data points
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Data Points")
plt.show()

# Step 3: Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)  # Initialize KMeans with 4 clusters
kmeans.fit(X)  # Fit the model to the data

# Step 4: Get cluster centers and labels
centers = kmeans.cluster_centers_  # Cluster centroids
labels = kmeans.labels_  # Labels for each data point

# Step 5: Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, cmap='viridis')  # Data points colored by cluster
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')  # Centroids
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

**Hirarichal clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.datasets import make_blobs

# Step 1: Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Step 2: Perform Hierarchical Clustering
# 'ward' minimizes variance between clusters, commonly used
linked = linkage(X, method='ward')

# Step 3: Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, truncate_mode='level', p=5)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

# Step 4: Create clusters
# Define the number of clusters
num_clusters = 4
labels = fcluster(linked, num_clusters, criterion='maxclust')

# Step 5: Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("Hierarchical Clustering")
plt.show()
```

**DBScan algorithm**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=30, cmap='viridis')
plt.title("Data points")
plt.show()

dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan.fit(X)

labels = dbscan.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.show()

unique_labels = set(labels)
n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
print(f"Number of clusters: {n_clusters}")
```

**Gaussian Mixture Model**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Step 1: Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Visualize the data points
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Data Points")
plt.show()

# Step 3: Apply Gaussian Mixture Model
gmm = GaussianMixture(n_components=4, random_state=42)  # Initialize GMM with 4 components
gmm.fit(X)  # Fit the model to the data

# Step 4: Get cluster labels and probabilities
labels = gmm.predict(X)  # Cluster labels for each data point
probs = gmm.predict_proba(X)  # Probabilities of each point belonging to each cluster
centers = gmm.means_  # Cluster centers (means of Gaussian components)

# Step 5: Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30, label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("Gaussian Mixture Model Clustering")
plt.legend()
plt.show()
```

**Markov Decision Process.**

```python
import numpy as np

# Step 1: Define MDP components
states = ["S1", "S2", "S3"]  # States
actions = ["A1", "A2"]  # Actions
transition_probabilities = {
    "S1": {"A1": {"S1": 0.5, "S2": 0.5}, "A2": {"S1": 0.2, "S3": 0.8}},
    "S2": {"A1": {"S2": 0.7, "S3": 0.3}, "A2": {"S1": 0.6, "S2": 0.4}},
    "S3": {"A1": {"S3": 1.0}, "A2": {"S1": 1.0}},
}
rewards = {
    "S1": {"A1": 5, "A2": 10},
    "S2": {"A1": 2, "A2": 7},
    "S3": {"A1": 0, "A2": 1},
}
gamma = 0.9  # Discount factor

# Step 2: Initialize value function
value_function = {state: 0 for state in states}

# Step 3: Policy evaluation
def policy_evaluation(policy, iterations=100, theta=1e-6):
    V = value_function.copy()
    for _ in range(iterations):
        delta = 0
        for state in states:
            v = V[state]
            action = policy[state]
            V[state] = sum(
                transition_probabilities[state][action][next_state] *
                (rewards[state][action] + gamma * V[next_state])
                for next_state in transition_probabilities[state][action]
            )
            delta = max(delta, abs(v - V[state]))
        if delta < theta:
            break
    return V

# Step 4: Define a policy
# Policy is a mapping of state -> action
policy = {
    "S1": "A2",  # In state S1, take action A2
    "S2": "A1",  # In state S2, take action A1
    "S3": "A2",  # In state S3, take action A2
}

# Step 5: Evaluate the policy
evaluated_values = policy_evaluation(policy)

# Step 6: Display results
print("Policy:")
for state, action in policy.items():
    print(f"  State {state}: Take action {action}")
print("\nValue Function:")
for state, value in evaluated_values.items():
    print(f"  State {state}: Value = {value:.2f}")
```

**Q-Learning algorithm**

```python
import numpy as np
import random

# Step 1: Define the environment
n_states = 6  # Number of states (e.g., grid positions)
n_actions = 2  # Actions: 0 = Left, 1 = Right
q_table = np.zeros((n_states, n_actions))  # Initialize Q-table with zeros
rewards = [-1, -1, -1, -1, -1, 10]  # Rewards for each state
gamma = 0.9  # Discount factor
alpha = 0.1  # Learning rate
epsilon = 0.1  # Exploration rate

# Step 2: Define the transition dynamics
def get_next_state(state, action):
    if action == 0:  # Move left
        return max(0, state - 1)
    elif action == 1:  # Move right
        return min(n_states - 1, state + 1)

# Step 3: Q-Learning algorithm
n_episodes = 500
for episode in range(n_episodes):
    state = random.randint(0, n_states - 2)  # Start in a random state, except the goal
    done = False

    while not done:
        # Choose an action using -greedy policy
        if random.uniform(0, 1) < epsilon:
            action = random.randint(0, n_actions - 1)  # Explore
        else:
            action = np.argmax(q_table[state])  # Exploit

        # Perform the action and observe the reward and next state
        next_state = get_next_state(state, action)
        reward = rewards[next_state]

        # Update Q-value using the Q-Learning formula
        q_table[state, action] += alpha * (
            reward + gamma * np.max(q_table[next_state]) - q_table[state, action]
        )

        # Transition to the next state
        state = next_state

        # If the agent reaches the goal, end the episode
        if state == n_states - 1:
            done = True

# Step 4: Display the learned Q-table
print("Learned Q-Table:")
print(q_table)

# Step 5: Test the learned policy
state = 0  # Start at the first state
path = [state]
while state != n_states - 1:
    action = np.argmax(q_table[state])
    state = get_next_state(state, action)
    path.append(state)

print("\nOptimal Path to Goal:")
print(" -> ".join(map(str, path)))
```

**Policy Gradient Method.**

```python
import numpy as np
import gym
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Step 1: Create the environment
env = gym.make("CartPole-v1")
n_actions = env.action_space.n  # Number of possible actions
n_states = env.observation_space.shape[0]  # Size of the state space

# Step 2: Build the Policy Network
def build_policy_network():
    model = Sequential([
        Dense(24, activation='relu', input_shape=(n_states,)),
        Dense(24, activation='relu'),
        Dense(n_actions, activation='softmax')  # Output probabilities for actions
    ])
    model.compile(optimizer=Adam(learning_rate=0.01), loss='categorical_crossentropy')
    return model

policy_network = build_policy_network()

# Step 3: Function to choose actions
def choose_action(state):
    state = state.reshape([1, n_states])  # Reshape for input to the model
    probabilities = policy_network(state).numpy()[0]
    action = np.random.choice(n_actions, p=probabilities)  # Sample action
    return action

# Step 4: Discounted reward computation
def compute_discounted_rewards(rewards, gamma=0.99):
    discounted_rewards = np.zeros_like(rewards, dtype=np.float32)
    cumulative = 0
    for t in reversed(range(len(rewards))):
        cumulative = cumulative * gamma + rewards[t]
        discounted_rewards[t] = cumulative
    return discounted_rewards

# Step 5: Training loop
def train_policy_gradient(episodes=500, gamma=0.99):
    for episode in range(episodes):
        state = env.reset()
        done = False
        states, actions, rewards = [], [], []

        while not done:
            action = choose_action(state)
            next_state, reward, done, _ = env.step(action)

            # Store trajectory
            states.append(state)
            actions.append(action)
            rewards.append(reward)

            state = next_state

        # Compute discounted rewards
        discounted_rewards = compute_discounted_rewards(rewards, gamma)
        discounted_rewards -= np.mean(discounted_rewards)  # Normalize rewards
        discounted_rewards /= np.std(discounted_rewards)

        # Convert actions to one-hot
        actions_one_hot = tf.keras.utils.to_categorical(actions, num_classes=n_actions)

        # Train policy network
        states = np.vstack(states)
        policy_network.train_on_batch(states, actions_one_hot, sample_weight=discounted_rewards)

        # Log progress
        total_reward = sum(rewards)
        print(f"Episode {episode + 1}/{episodes} - Total Reward: {total_reward}")

# Step 6: Train the agent
train_policy_gradient(episodes=500)

# Step 7: Test the trained agent
state = env.reset()
done = False
total_reward = 0

print("\nTesting the trained agent...")
while not done:
    env.render()
    action = choose_action(state)
    state, reward, done, _ = env.step(action)
    total_reward += reward

env.close()
print(f"Total reward achieved during testing: {total_reward}")
```
